```python
import pandas as pd
import numpy as np
```

```python
#Let's try with street names that contain the seasons
#and let's use the big GBG dataset

GBG_all = pd.read_csv("GBG all highways simplified.csv")
```

```python
#Get a smaller dataset that only contains relevant columns

GBG_subset=GBG_all[["@id", "highway", "name", "highway", "FID"]]
GBG_subset[:5]
```

```python
#Load the csv with the GBG_cluster "culture"
GBG_cluster_culture = pd.read_csv("GBG_cluster_culture.csv")
GBG_cluster_culture.head()
```

```python
#Make a list out of this column
GBG_cluster_culture_list = GBG_cluster_culture["culture"].tolist()
print(GBG_cluster_culture_list)
```

```python
#Create a list of our conditions
culture_conditions_list = []
for item in GBG_cluster_culture_list:
    culture_conditions_list.append((GBG_subset["name"].str.startswith(item) == True))

#print (culture_conditions_list)
```

```python
#Create a list of the values we want to assign for each condition
#We want to actually give all those streets the value "culture", so we need a list of the same number
#of items
culture_values = []
for item in GBG_cluster_culture_list:
    culture_values.append("culture")

print(len(culture_values))
print(len(GBG_cluster_culture_list))
```

```python
#Now, create a new column and use np.select to assign values to it using our list as arguments
#GBG_subset["category"] = np.select(list of conditions, category values, "other")
#We don't want to fill in "other", since this will be filled with other categories later

GBG_subset["category"] = np.select(culture_conditions_list, culture_values)
```

```python
GBG_subset.to_csv("GBG_subset with category.csv")
```

## loc
loc is for filtering rows and columns **by label**, i.e. for rows by index, for columns by column names.

df.loc[what rows do I want, what columns do I want]

For row 0 and all columns: df.loc[0, :]

You can also pass a list in, e.g. for rows 0, 3 and 7: df.loc[[0,3,7], :]

For continuous rows: df.loc[0:3, :] --> inclusive on both sides

For specific columns by name: df.loc[: , ["city", "state"]]









### loc with Boolean conditions

E.g. if you want to select all rows where city = Oakland: df[df.city == "Oakland]

But the same thing can be done with .loc:

df.loc[ufo.city == "Oakland, :]

Advantage with the .loc method: it will not cause problems related to chained indexing!




## iloc

iloc is for filtering rows and columns **by integer position**

e.g. all rows and columns in position 0 and 3: df.iloc[:, [0, 3]]

For continuous columns: df.iloc[:, [0:3]] --> exclusive on the right side!

Apply is both a series and a dataframe method. It applies a function to each element in a series or dataframe.

```python
gbg = pd.read_csv("Unique GBG street names.csv")
gbg.head(20)
```

```python
radio_cluster = pd.read_csv("cluster_radio.csv")
culture_cluster = pd.read_csv("cluster_culture.csv")
space_cluster = pd.read_csv("cluster_space.csv")
professions_cluster = pd.read_csv("cluster_professions.csv")
colors_cluster = pd.read_csv("cluster_colors.csv")
animals_cluster = pd.read_csv("cluster_animals.csv")
industrial_cluster = pd.read_csv("cluster_industrial.csv")
doctors_cluster = pd.read_csv("cluster_doctors.csv")
food_cluster = pd.read_csv("cluster_food.csv")
weather_cluster = pd.read_csv("cluster_weather.csv")
holidays_cluster = pd.read_csv("cluster_holidays.csv")
flora_cluster = pd.read_csv("cluster_flora.csv")
gems_cluster = pd.read_csv("cluster_gems.csv")
radio_cluster.head()
```

```python
#Turn into list
culture_list = culture_cluster["culture"].tolist()
radio_list = radio_cluster["radio"].tolist()
space_list = space_cluster["space"].tolist()
professions_list = professions_cluster["professions"].tolist()
colors_list = colors_cluster["colors"].tolist()
animals_list = animals_cluster["animals"].tolist()
industrial_list = industrial_cluster["industrial"].tolist()
doctors_list = doctors_cluster["doctors"].tolist()
food_list = food_cluster["food"].tolist()
weather_list = weather_cluster["weather"].tolist()
holidays_list = holidays_cluster["holidays"].tolist()
flora_list = flora_cluster["flora"].tolist()
gems_list = gems_cluster["gems"].tolist()
```

```python
def check_category(row):
    if row['street names'] in culture_list:
        cat = "culture"
    elif row['street names'] in radio_list:
        cat = "radio"
    elif row['street names'] in space_list:
        cat = "space"
    elif row['street names'] in professions_list:
        cat = "professions"
    elif row['street names'] in colors_list:
        cat = "colors"
    elif row['street names'] in animals_list:
        cat = "animal"
    elif row['street names'] in industrial_list:
        cat = "industrial"
    elif row['street names'] in doctors_list:
        cat = "doctors"
    elif row['street names'] in food_list:
        cat = "food"
    elif row['street names'] in weather_list:
        cat = "weather"
    elif row['street names'] in holidays_list:
        cat = "holidays"
    elif row['street names'] in flora_list:
        cat = "flora"
    elif row['street names'] in gems_list:
        cat = "gems"
    else:
        cat = "other"
    return cat
```

```python
gbg['category'] = gbg.apply(check_category, axis=1)
```

```python
gbg_new = gbg.rename(columns={"street names":"name"})
```
